Scrape data from wiki
Good read on how to read html to df: 
https://beenje.github.io/blog/posts/parsing-html-tables-in-python-with-pandas/



```python
#The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood
toronto_wiki = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
raw_toronto_wiki_page = requests.get(toronto_wiki)
soup = BeautifulSoup(raw_toronto_wiki_page.content,'lxml')
areas = soup.find_all('table')[0]
df = pd.read_html(str(areas), header=0)[0]
df.head()
```

```python
#Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
df = df[df.Borough != 'Not assigned']
df.head()
```

```python
#More than one neighborhood can exist in one postal code area. 
# These rows will be combined into one row with the neighborhoods separated with a comma

df.groupby('Postcode')['Neighbourhood'].apply(','.join).reset_index()
```
